```python
import face_recognition
import os
import cv2

KNOWN_FACES_DIR='FaceRecG1/known_faces'
UNKNOWN_FACES_DIR='FaceRecG1/unknown_faces'
TOLERANCE=0.6
MODEL = 'hog' # 'hog' or 'cnn'

REGTANG_RGBCOLOR = [0, 255, 0]
```

```python
print('loading knwon faces...')
known_faces = []
known_names = []

for name in os.listdir(KNOWN_FACES_DIR):
    for filename in os.listdir(f'{KNOWN_FACES_DIR}/{name}'):
        image = face_recognition.load_image_file(f'{KNOWN_FACES_DIR}/{name}/{filename}')
        encoding128d = face_recognition.face_encodings(image)[0]
        known_faces.append(encoding128d)
        known_names.append(name)
        
print(known_faces)
print(known_names)
```

```python
print('loading unknwon faces...')
# unknown_faces = []
# unknown_names = []

for filename in os.listdir(UNKNOWN_FACES_DIR):
    image = face_recognition.load_image_file(f'{UNKNOWN_FACES_DIR}/{filename}')
    locations = face_recognition.face_locations(image, model=MODEL)
    encoding128d = face_recognition.face_encodings(image, locations)
    image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

    for face_encoding, face_location in zip(encoding128d, locations):
        results = face_recognition.compare_faces(known_faces, face_encoding, TOLERANCE)
        # print(results)
    
        match = None
        if True in results:
            match = known_names[results.index(True)]
            print(f' - {match} from {results}')
            top_left = (face_location[3], face_location[0])
            bottom_right = (face_location[1], face_location[2])
            color = REGTANG_RGBCOLOR
            cv2.rectangle(image, top_left, bottom_right, color, 3)
            
            top_left = (face_location[3], face_location[2])
            bottom_right = (face_location[1], face_location[2]+22)
            cv2.rectangle(image, top_left, bottom_right, color, cv2.FILLED)
            font_size = 0.5
            font_color = (200,200,200)
            cv2.putText(image, match, (face_location[3]+ 10,face_location[2] + 15), cv2.FONT_HERSHEY_SIMPLEX, font_size, font_color, 2)
            
    cv2.imshow(filename, image)
    cv2.waitKey(5000)
```
